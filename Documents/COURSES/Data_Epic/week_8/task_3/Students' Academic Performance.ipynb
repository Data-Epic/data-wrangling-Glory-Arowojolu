```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("student-scores.csv")
df.head()
```

```python
df.info()
```

## Data Preprocessing and Cleaning

```python
df.nunique()
```

```python
print(df.duplicated().sum())
```

```python
df.isnull().sum()
```

```python
# Drop irrelevant columns
df = df.drop(columns=['id', 'first_name', 'last_name', 'email', 'gender', 'career_aspiration'])
df.head()
```

## Exploratory Data Analysis (EDA)

```python
df.describe()
```

```python
# create a new column named Overall performance to get the aggregate score of each student across different subjects
subject_scores = ['math_score', 'history_score', 'physics_score', 'chemistry_score', 'biology_score', 'english_score', 'geography_score']

df['overall_performance'] = df[subject_scores].mean(axis=1)
df['overall_performance'] = df['overall_performance'].astype(int)
df.head()
```

### 1) # How do weekly self-study hours impact students' scores across different subjects?

```python
# Group by weekly study hours and compute the mean of each subject
df_avg = df.groupby('weekly_self_study_hours')[subject_scores].mean().reset_index()

# then, reshape the major columns involved here into a long format
df2 = df_avg.melt(id_vars=['weekly_self_study_hours'], value_vars=subject_scores, var_name='Subject', value_name='Average Score')
print(df2.head())
```

```python
# Create a scatterplot with trendline
sns.lmplot(data=df2, x='weekly_self_study_hours', y='Average Score', hue='Subject', height=6, aspect=1.5, scatter_kws={'alpha': 0.6}, ci=None)
plt.xlabel('Weekly Self-Study Hours')
plt.ylabel('Subject Scores')
plt.title('Impact of Weekly Self-Study on Students scores')
plt.show()
```

### 2) Do students with part-time jobs perform differently compared to those without jobs?

```python
# Group by part_time_job and compute the mean of overall performance
df_grp = df.groupby('part_time_job')['overall_performance'].mean().reset_index()

# Create a bar chart
sns.barplot(data=df_grp, x='part_time_job', y='overall_performance')
plt.xlabel('Part-Time Job')
plt.ylabel('Average Overall Performance')
plt.title('The Effect of Part-time job status on Students performance')
plt.show()
```

### 3) Is there a relationship between absence days and lower academic performance?

```python
# Create a scatterplot with trendline

sns.lmplot(data=df, x='absence_days', y='overall_performance', ci=None)
plt.xlabel('Absence days')
plt.ylabel('Overall Performance')
plt.title('Overall Performance vs Absence Days')
plt.show()
```

### 4) Do extracurricular activities help or hurt student performance?

```python
# Create a boxplot
sns.boxplot(data=df, x='extracurricular_activities', y='overall_performance')
plt.xlabel('Extracurricular Activities')
plt.ylabel('Overall Performance')
plt.title('Extracurricular activities effect on Students performance')
plt.show()
```

### 5) Which subject has the strongest correlation with overall academic success?

```python
# Selecting all rows but the last 8 columns which consist of different subject scores and the overall performance column
df3 = df.iloc[:, -8:]

df3.head()
```

```python
# Create a correlation heatmap
correlation_matrix = df3.corr()
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix of Subject Scores')
plt.show()
```

## Insights & Interpretation

1. The scatter plot shows a positive correlation between weekly self-study hours and students' scores across different subjects. This implies that as the number of hours spent weekly to study increases, the scores also increase in various subjects.
2. From the bar chart, it's observed that the average overall performance of students without jobs is slightly higher than those who work part-time. This indicates that the involvement of students in part-time jobs has a relatively lower effect on their academic performance.
3. The scatter plot shows a negative correlation between absence days and overall academic performance. The downward trend in the scatter plot implies that as the number of absence days increases, the students' academic performance declines.
4. The boxplot compares students that engage in extracurricular activities to those who do not in regards to their academic performance. Both categories have a median overall performance of 82, this implies that being involved in extracurricular activities doesn't hurt their academic performance. The interquartile range (IQR) of both categories are also similar, with 50% of the students scoring within the range of 77 and 84.
5. The correlation matrix reveals that math has the highest correlation (0.51) with overall academic performance, compared to other subjects. This implies that students with higher scores in math  are more likely to have a higher overall academic success. Other subjects like biology(0.5), history(0.49) and chemistry(0.49) have moderate correlations but not as strong as math.
